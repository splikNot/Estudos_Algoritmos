```python
"""
o Algoritmo deve:
 - Fazer a LCS entre X e Y
 - Calcular o Comprimento da maior subsequencia
 - Gerar uma subsequencia de comprimento L
"""

def lcs(X, Y):
  m = len(Y) #Linhas  #Pegando as dimensôes da string
  n = len(X) #Colunas

  grid = [[0 for j in range (n + 1)] for i in range (m + 1)]

  for i in range(1, m+1): #Varrendo as linhas a frente do zero "fonte"
    for j in range(1, n+1): #O mesmo nas colunas
      if Y[i-1] == X[j -1]: # Se os valores forem iguais
        grid[i][j] = grid[i-1][j-1] + 1 #Somamos 1 àquela posição
      else: #Caso contrário
        grid[i][j] = max(grid[i][j-1],grid[i-1][j]) # Aquela posição recebe o maior valor entre as suas diagonais

  
  #Reconstruindo a sequencia a partir do grid

  L = grid[m][n]     #Aqui não precisa alterar o indice, devido a propriedade dos indices iniciarem em 0
  seq = [""]*(L+1)   #Criando uma lista agora que já sei o tamanho da sequencia (Com uma posição a mais) o porque eu vou descobrir
  seq[L] = ""

  i = m
  j = n
  while i > 0 and j > 0: #Lendo a tabela do canto direito inferior ao canto esquerdo superior
    if Y[i-1] == X[j-1]: #Se o número na vertical ou horizontal é igual
      seq[L-1] = Y[i-1] # Guardamos o valor daquela posição
      i -= 1    
      j -= 1            #Reduzimos todos os indices
      L -= 1

    elif grid[i-1][j] > grid[i][j-1]: #Se o valor de cima for maior que o valor a esquerda
      i -= 1  #A gente sobe
    else:
      j -= 1 #Caso contrário a gente continua a varredura pela linha
  

  return (grid, "".join(seq[:-1]), str(grid[m][n])) #Como a gente salvou uma posiçao a mais em Seq, a gente tira aquele espaço vazio agora


```

```python
"""Função para a leitura dos arquivos e concatenação do código genético em uma única string"""

def LeArquivoDNA(filename):
  files=open(filename,'r')
  lists=files.readlines() #this is a matrix of nlines
  nlines=len(lists) # number of lines

  a = lists[0].rstrip('\n').split(' ') # separa a linhas em colunas
                                        # observa o espaco ’ ’ como
                                          # caracter de separacao
                                              # descarta \n

  cadeiacompleta = a[1] + a[2] + a[3] + a[4] + a[5] + a[6] #Cada linha está separada em 6 cadeias com indice na posição 0 que serve como referencia a posição do
                                                          #Aminoácido no inicio de cada linha
                                                          # concatena as colunas 1..6 ignora a[0]
                                                        # agora que cadeiacompleta nao é, vazio faca ate o final
  for i in range(1, nlines):
    a=lists[i].rstrip('\n').split(' ')
    cadeiacompleta = cadeiacompleta + a[1]+a[2]+a[3]+a[4]+a[5]+a[6]
#retorna a string completa
  return(cadeiacompleta)
```
